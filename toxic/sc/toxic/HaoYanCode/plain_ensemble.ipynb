```python
import os


SUB_DIR = '/home/kai/data/shiyi/toxic/submissions/'
count = 0
ensemble_id = ''
final = ''
for root, dirs, files in os.walk(SUB_DIR):
    for filename in files:
        if '_50_50_0.3_32_' in filename or '_None_' in filename:
            ensemble_id += str(filename.split('.')[-2].split('_')[-1] + '_')[-5:]
            sub_file = SUB_DIR + filename
            if count == 0:
                final = pd.read_csv(sub_file)[label_cols]
            else:
                final += pd.read_csv(sub_file)[label_cols]
            count+=1
    
print(count)
final = final / count
submission = pd.read_csv(SUB_DIR + 'sample_submission.csv')
submission[label_cols] = final
submission.to_csv(SUB_DIR + ensemble_id + '.csv', index=False)
    
print('done')
```
