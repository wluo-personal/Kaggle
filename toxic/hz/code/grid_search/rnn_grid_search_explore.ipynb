```python
import os
subres = {}
for filename in os.listdir('../../model'):
    if 'csv' in filename:
        if filename.startswith('9'):
            score = filename[:4]
            csvname = filename[4:]
            subres[csvname] = int(score)
            print(score, csvname)
```
