```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python

svm_1 = pd.read_csv('../data/nblogreg_best_nbappearance_c_0.2.csv')

log_1 = pd.read_csv('../data/nblogreg_best_c_0.2.csv')
```

```python
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

df = {col: pd.DataFrame() for col in label_cols}

for col in label_cols:
    df[col]['svm_1'] = svm_1[col]
    df[col]['log_1'] = log_1[col]
```

```python
scatter_matrix(df['toxic'])
plt.show()
```
