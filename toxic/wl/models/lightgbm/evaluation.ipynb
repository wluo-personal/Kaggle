```python
import pandas as pd
file = '/home/kai/data/kaggle/toxic/wl/models/data/lgb_all_gridsearch.csv'
df_lgb = pd.read_csv(file)

label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']


best_params = pd.DataFrame({col:[] for col in df_lgb.columns})
 
for col in label_cols:
        tmp = df_lgb[(df_lgb['trained'] == 'Y') & (df_lgb['label'] == col) ]
        best_params = pd.concat([best_params,pd.DataFrame( tmp.sort_values(['roc'], ascending=False).reset_index(drop=True).loc[0]).transpose()])  
```

```python
### get 
appen1 = df_lgb[ df_lgb['max_depth'] == 3].copy()
appen2 = df_lgb[ df_lgb['max_depth'] == 3].copy()
```

```python
appen1['max_depth'] = appen1.max_depth.apply(lambda x: 6)
appen1['roc'] = appen1.roc.apply(lambda x: 0)
appen1['trained'] = appen1.trained.apply(lambda x: 'N')


appen2['max_depth'] = appen2.max_depth.apply(lambda x: -1)
appen2['roc'] = appen2.roc.apply(lambda x: 0)
appen2['trained'] = appen2.trained.apply(lambda x: 'N')
```

```python
df_lgb.shape
```

```python
df_lgb1 = pd.concat([df_lgb, appen1, appen2])
```

```python
df_lgb1.shape
```

```python
df_lgb1.to_csv(PATH + 'lgb_all_gridsearch.csv', index=False)
```
