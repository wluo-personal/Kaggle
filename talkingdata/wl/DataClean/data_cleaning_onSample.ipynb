```python
import multiprocessing

def _apply_df(args):
    df, func, kwargs = args
    return df.apply(func, **kwargs)

def _build_in_df(args):
    df, kwargs = args
    time_type = kwargs.pop('time_type')
    if time_type == 'year':
        return df.year.astype('uint16')
    elif time_type == 'month':
        return df.month.astype('uint8')
    elif time_type == 'week':
        return df.week.astype('uint8')
    elif time_type == 'day':
        return df.day.astype('uint16')
    elif time_type == 'minute':
        return df.minute.astype('uint8')
    elif time_type == 'second':
        return df.second.astype('uint8')

def apply_by_multiprocessing(df, func, **kwargs):
    workers = kwargs.pop('workers')
    pool = multiprocessing.Pool(processes=workers)
    result = pool.map(_apply_df, [(d, func, kwargs)
            for d in np.array_split(df, workers)])
    pool.close()
    return pd.concat(list(result))

def build_in_by_multiprocessing(df, **kwargs):
    workers = kwargs.pop('workers')
    pool = multiprocessing.Pool(processes=workers)
    result = pool.map(_build_in_df, [(d, kwargs)
            for d in np.array_split(df, workers)])
    pool.close()
    return pd.concat(list(result))

def func(x):
    return x.timestamp()

aa = apply_by_multiprocessing(clicks, func, workers=10)
```
