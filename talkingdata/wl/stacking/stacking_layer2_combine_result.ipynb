```python
import pandas as pd
import lightgbm as lgb
import numpy as np
```

```python
entry = 5
K = 3
folds = [0, 1, 2]
folds_comp = [[1, 2], [0, 2], [0, 1]]
combine = 0

base_path = '/home/kai/data/kaggle/talkingdata/wl/data/stacking/train/result/'
base_path_test = '/home/kai/data/kaggle/talkingdata/wl/data/stacking/test/result/'

func_pool = ['count', 'mean', 'reversemean', 'time2nextclick', 
             'time2previousclick', 'countfromfuture', 'countfrompast', 'lasttimediff']

func_pool = ['count']

filebase = 'train_fold{}_predonfold{}_{}_{}_param{}.npy'
filebase_test = 'test_fold{}_{}_{}_param{}.npy'

# filebase = 'train_fold{}_predonfold{}_{}_{}_param{}.npy'
# filebase.format(0,1,'count',0,0)
```

# Combine Train

```python
def generate_layer2_features_train(func_pool, entry, folds, folds_comp, base_path, filebase,combine=0):
    for func in func_pool:
        for ent in range(entry):
            result_list_underentry = []

            for on_pred in folds:
                each_fold_in_result = []
                for comp_fold in folds_comp[on_pred]:
                    pred_file_name = base_path + filebase.format(comp_fold,on_pred, func, ent, combine)
                    print('loading file: \n{}'.format(pred_file_name))
                    pred = np.load(pred_file_name)
                    each_fold_in_result.append(pred)
                result_list_underentry.append(np.mean(each_fold_in_result, axis=0))
                print('averaging on fold {} done!'.format(on_pred))
            result = np.concatenate(result_list_underentry)
            save_name = 'train_layer2_{}_{}_param{}.npy'.format(func, ent, combine)
            save_path = base_path + 'layer2/' + save_name
            np.save(save_path, result)
            print('concate these folds')
            print(len(result))
            print(save_path)
            print('=========================================')
generate_layer2_features_train(func_pool, entry, folds, folds_comp, base_path, filebase)
```

# Combine Test

```python
def generate_layer2_features_test(func_pool, entry, folds, folds_comp, base_path, filebase, combine=0):
    for func in func_pool:
        for ent in range(entry):
            result_list_underentry = []

            for on_pred in folds:
                each_fold_in_result = []
                pred_file_name = base_path + filebase.format(on_pred, func, ent, combine)
                print('loading file: \n{}'.format(pred_file_name))
                pred = np.load(pred_file_name)
                each_fold_in_result.append(pred)
                print('adding fold {}...'.format(on_pred))
            result = np.mean(each_fold_in_result, axis=0)
            print('averaging done!'.format(on_pred))
            save_name = 'test_layer2_{}_{}_param{}.npy'.format(func, ent, combine)
            save_path = base_path + 'layer2/' + save_name
            np.save(save_path, result)
            print(len(result))
            print(save_path)
            print('=========================================')
generate_layer2_features_test(func_pool, entry, folds, folds_comp, base_path_test, filebase_test)
```
