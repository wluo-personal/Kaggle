```python
import re

def dict_return(col_idx, file):
    """
    d is the dictionary of count
    d1 is the dicitonary of count is_attribute=1
    """
    col = file.readline()
    line = file.readline()
    it = 1
    d = {}
    d1 = {}
    while line:
        vline = line.split(',')
        vline[-1] = re.sub(r'\n', '', vline[-1])
        tmp = vline[col_idx]
        if not d.__contains__(tmp):
            d[tmp] = 0
            d1[tmp] = 0
        d[tmp] += 1
        if int(vline[-1]) == 1:
            d1[tmp] += 1
        line = file.readline()
        it += 1
        if it % 10000000 == 0:
            print(it, len(d), len(d1))
    return d, d1

print('done')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def figure_plot(d, d1):
    """
    d is the dictionary of count
    d1 is the dicitonary of count is_attribute=1
    """
    dt = {}
    for x in d.keys():
        dt[x] = d1[x] / d[x]

    x = sorted([int(k) for k in dt.keys()])
    y = []
    for l in x:
        y.append(dt[str(l)])

    fig = plt.figure(figsize=(20,10))
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x,y, '-ob')
    ax.grid()
    ax.set_yticks(np.arange(0, 1.1, 0.1))
    plt.show()
```

# app

```python
import json
from time import time
from pathlib import Path

col_name = 'app'
PATH = '../data/'

t0 = time()
if Path(PATH + col_name + '.json').is_file():
    with open(PATH + col_name + '.json', 'r') as f:
        data = json.load(f)
        d = data['d']
        d1 = data['d1']
else:
    with open(PATH + 'train_cleaned_final.csv', 'r') as f:
        d, d1 = dict_return(1, f)
    with open(PATH + 'app.json', 'w') as f:
        json.dump({'d':d, 'd1':d1}, f)
        
print(time() - t0)
```

```python
figure_plot(d, d1)
```
