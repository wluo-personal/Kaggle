```python
PATH = '../data/'

f = open(PATH + 'train_cleaned_final.csv', 'r')
col = f.readline()
line = f.readline()

current = '7'
g = open(PATH + current + '.csv', 'a')
print(PATH + current + '.csv')
g.write(col)
iter = 1
while line:
    lines = line.split(',')
    if iter % 1000000 == 0:
        print(iter, lines[5])
    iter += 1
    if lines[5] == current:
        g.write(line)
    if lines[5] == str(int(current) + 1):
        current = str(int(current) + 1)
        g.close()
        g = open(PATH + current + '.csv', 'a')
        print(PATH + current + '.csv')
        g.write(col)
        g.write(line)
    line = f.readline()
```

```python
PATH = '../data/'
h4 = []
h5 = []
h6 = []
h9 = []
h10 = []
h11 = []
h13 = []
h14 = []
h15 = []
def distr(c):
    with open(PATH + str(c) + '.csv') as f:
        col = f.readline()
        line = f.readline()
        iter = 1
        while line:
            if iter % 10000000 == 0: print(iter)
            iter += 1
            lines = line.split(',')
            if lines[6] == '4': h4.append(line)
            elif lines[6] == '5': h5.append(line)
            elif lines[6] == '6':
                if lines[8] == '0' and lines[9] == '0':
                    h6.append(line)
            elif lines[6] == '9': h9.append(line)
            elif lines[6] == '10': h10.append(line)
            elif lines[6] == '11':
#                 print(lines)
                if lines[8] == '0' and lines[9] == '0':
                    h11.append(line)
            elif lines[6] == '13': h13.append(line)
            elif lines[6] == '14': h14.append(line)
            elif lines[6] == '15':
                if lines[8] == '0' and lines[9] == '0':
                    h15.append(line)
            line = f.readline()
print('done')
```
